# Question 1. Demographic:

a. Which 2 demographics have proven to be the most popular with the telco brand?

```python
Churn_No= df[df["Churn"]== "No"]
Churn_Yes= df[df["Churn"]== "Yes"]
```

```python
Churn_No_df= Churn_No.groupby(["gender", "SeniorCitizen", "Partner", "Dependents"])["customerID"].count().reset_index(name= "Count")
Churn_No_df
```

```python
sns.catplot(x='Dependents', y='Count', row ='Partner', col = 'SeniorCitizen', hue = 'gender', data= Churn_No_df, kind= "bar")
```

Young citizens with no partners and no dependents, Young citizens with partners and dependents

b. If the marketing team wanted to increase customer retention, to whom should they target their advertisements to?

```python
Churn_Yes_df= Churn_Yes.groupby(["gender", "SeniorCitizen", "Partner", "Dependents"])["customerID"].count().reset_index(name= "Count")
Churn_Yes_df
```

```python
sns.catplot(x='Dependents', y='Count', row ='Partner', col = 'SeniorCitizen', hue = 'gender', data= Churn_Yes_df, kind= "bar")
plt.show()
```

Young citizens with no partners and no dependents

# 2. Services:

a. Which 3 services are contributing to a higher monthly charge and resulting to customers churning? (Use visualizations in explaining your analysis)

```python
services = ['PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for service in services:
    sns.barplot(x= service, y= "MonthlyCharges", data= df , hue = 'Churn')
    plt.title(service)
    plt.grid()
    plt.show()
```

b. If the telco was to offer Phone as a standalone service, which type of contract would encourage customer retention?

```python
phone_service_df= df[df['PhoneService'] == 'Yes']
phone_service_df.head()
```

```python
sns.countplot(x="Contract", hue="Churn", data= phone_service_df)
plt.show()
```

Because very few customers churn inÂ this category, the telco would be in an advantageous position to offer a two-year contract.

# 3. Payment: (Use visualizations in explaining your analysis)

a. If the company was to streamline all its services into 3 bouquet packages, what monthly prices will be appropriate for the following packages to keep customers from churning:
i. Basic
ii. Plus
iii. Premium

```python
x=Churn_No['MonthlyCharges']
Churn_No['MonthlyCharges_Range'] = pd.cut(x, bins=3, labels=["Basic","Plus", "Premium"])
Churn_No.head()
```

```python
Churn_No.groupby("MonthlyCharges_Range")['MonthlyCharges'].median()
```

According to the visualization above, we would accord the prices as follows:

 i. Basic - $20.8, 
 ii. Plus - $70.4, 
 iii. Premium - $99.0
